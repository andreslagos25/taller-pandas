```python
# Lectura de archivos XML
"""
    Para la lectura de archivos XML se necesita minimo dos cosas
    un XPATH y un prefix namespace(diccionario)
    XPATH='ruta' El XPATH permite recorrer el XML para buscar dentro de las 
    diferentes etiquetas que pueden contener la informacion
    namespaces=diccionario Permite identificar que atributo especifico recorrer
    Dado que un XML puede contener muchos atributos con el mismo nombre
    si tiene un namespace diferente puede ser filtrado
"""
namespaces = {
    'gesmes': 'http://www.gesmes.org/xml/2002-08-01',
    'eurofxref': 'http://www.ecb.int/vocabulary/2002-08-01/eurofxref'
}
pd.read_xml("../data/data.xml", xpath='.//eurofxref:Cube/eurofxref:Cube/eurofxref:Cube', namespaces=namespaces).head(5)
```
